## Fillna

```python
import pandas as pd
var = pd.read_csv("Lec 09.csv")
var

# Here u can see various Missing field denoted by NaN we will work with this 
```

```python
# u can fill it with any value you want 
var.fillna("no value")
```

```python
# If u want to fill column specific values u can pass dictionary
var.fillna({"First Name":"AAAAA", "Last Name":"BBBBB", "Salary":"00000", "Age":0})
```

```python
# If u want to fill the forward data in null
var.fillna(method = "ffill")
```

```python
# If u want to fill the backward data in null
var.fillna(method = "bfill")
```
