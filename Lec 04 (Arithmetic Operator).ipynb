## Arithmetic Operator

```python
import pandas as pd
# Create a DataFrame
x = {"A":[1, 2, 3, 4, 5],"B":[6, 7, 8, 9, 10]}

var = pd.DataFrame(x)

var
```

```python
# Addition : while performing this operation u need to make a new column and result is stored in it 
var["C"] = var["A"] + var["B"]

var
```

```python
# similarly u can perform various arithmateic operation 

# Subtraction
var["D"] = var["A"] - var["B"]

# Multiplication
var["E"] = var["A"] * var["B"]

# Division
var["F"] = var["A"] / var["B"]

var
```

```python
# U acn also perform logical operation and get boolean answer 

# Create a DataFrame
y = {"A":[10, 20, 30, 40, 50],"B":[16, 17, 18, 19, 10]}
var1 = pd.DataFrame(y)
var

# Comparing
var1["Greater"] = var1["A"] <= 20
var1

# Modulus
var1["Even"] = var1["B"] % 2 == 0
var1
```
