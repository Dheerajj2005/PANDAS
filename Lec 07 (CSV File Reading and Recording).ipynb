## CSV File Reading and Recording

```python
import pandas as pd

csv = pd.read_csv("username.csv")

csv
```

```python
# If u want particular rows 
csv1 = pd.read_csv("username.csv",nrows=3)

csv1
#print(csv1.columns)
```

```python
# If u want particular column 
csv2 = pd.read_csv("username.csv",usecols = ["Identifier", "Last name"])
csv2

# Be carefull while entering column name here there is spaces before and after identifier
```

```python
# u can aslo do this using column index number 
csv3 = pd.read_csv("username.csv",usecols = [1,3])
csv3
```

```python
# If u want to skip any rows use skiprows() function
csv4 = pd.read_csv("username.csv",skiprows = [1,3])
csv4
```

```python
# Here numbers are index if u want to make any column as index use index_col function 
csv5 = pd.read_csv("username.csv",index_col = "First name")
csv5
```

```python
csv6 = pd.read_csv("username.csv")
csv6
```

```python
# If u want to make any row its header 
csv7 = pd.read_csv("username.csv", header = 2)
csv7
```

```python
# If u want to give more heading 
csv8 = pd.read_csv("username.csv", names = ["cols1","cols2","cols3","cols4","cols5"])
csv8
```

```python
# If u don't want any header
csv9 = pd.read_csv("username.csv", header = None)
csv9
```

```python
# If u want to change the datatype
csv_1 = pd.read_csv("username.csv", dtype = {"Salary":"float"})
csv_1
```
