# Dropna and Fillna  

```python
import pandas as pd
var = pd.read_csv("Lec 09.csv")
var

# Here u can see various Missing field denoted by NaN we will work with this 
```

```python
# If u want to drop all the missing field use dropna
var.dropna()
# Here drop as occured along row that is axis = 0 
# for column drop use axis = 1 
```

```python
var.dropna(axis = 1)
```

```python
# If u see at row 3 whole row is empty to remove this we use 
var.dropna(how = "all")
# While using all u can see other NaN fields so to not see other NaN fiels you can use any 
```

```python
var.dropna(how = "any")
```

```python
# If u want to delete NaN along any column u can use subset
var.dropna(subset=["Last Name"])
```

```python
# If u want to drop a row with a single or double or any number null value u can 
var.dropna(thresh = 1)
var
```
