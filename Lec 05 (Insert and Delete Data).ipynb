## Insert and Delete Data

```python
import pandas as pd
```

```python
# Create a DataFrame
var = pd.DataFrame({"A":[1, 2, 3, 4, 5],"B":[6, 7, 8, 9, 10]})
var
```

**insert** <br>
*There are 3 Parameter* <br>
1) index -- what index u want to insert <br>
2) position and name <br>
3) Data to be inserted <br>

```python
var.insert(1, "C", var["A"])

var
```

```python
var.insert(3, "B*A", var["A"] * var["B"])

var
```

```python
# Passing a list 
var.insert(4, "D", [5, 6, 7, 8, 9])

var

# Note : Enter number of proper length otherwise error will occur
```

```python
# If u want to copy data of particular row uo=pto certain value 
# Use slicing 

var["Copy"] = var["A"][:3]

var
```

# Delete

```python
import pandas as pd

# create a DataFrame
var1 = pd.DataFrame({"A":[1, 2, 3, 4, 5],"B":[6, 7, 8, 9, 10],"C":[6, 4, 3, 8, 7]})
var1
```

```python
# Delete 
x = var1.pop("B")
# Print the deleted column
print(x)
```

```python
# Printing the new updated DataFrame
var1
```

```python
# Can also used del function 
del var1["A"]
var1
```
