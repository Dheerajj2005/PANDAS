**What is pandas ?**
1) The name "pandas" has a reference to both "panel data", and "python data analysis" and was created by wes mckinney in 2008. <br>
2) Pandas is a python library used for working with data sets. <br>
3) It has functions for analyzing, cleaning, exploring, and manipulating data. <br>
4)  Read and write data structures and different formats : csv, JSON,ZIP etc. <br>

**Pandas data structure** 
* There are 3 data structure <br>
1) Series: One-Dimensional labeled arrays pd.Series(data) <br>
2) DataFrames: Two-Dimensional data structure with columns, much like a table. <br>
3) Panel : A panel is a 3D container of data. <br>

**Importance of Pandas in Python** 

1) Pandas allows us to analyze big data and make conclusions based on statistical theories.<br> 
2) Pandas can clean messy data sets, and make them readable and relevant.<br>
3) Easy handling of missing data (represented as NaN) in floating point as well as non-floating point data. <br>
4) Size mutability: columns can be inserted and deleted from DataFrame and higher dimensional objects <br>
5) Data set merging and joining. Flexible reshaping and pivoting of data sets Proftime-- series functionality. <br>
