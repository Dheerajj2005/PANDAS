## Replace and Interpolate

```python
import pandas as pd
var = pd.read_csv("Lec 09.csv")
var

# Here u can see various Missing field denoted by NaN we will work with this 
```

```python
var.replace(to_replace = "Dheeraj", value = "Yadav" )
```

```python
var.replace([19, 20, 21, 23, 15, 20, 25], 50)
```

```python
# Second way to replace use regex 
# replace capital A to Z and small a to z
var.replace("[A-Za-z]", "dheeraj", regex = True )
```

```python
# Only particular column 
var.replace({"First Name":"[A-Z]"},22, regex = True)
```

```python
var.replace(20,method = "bfill") 
# similarly for ffill
```
